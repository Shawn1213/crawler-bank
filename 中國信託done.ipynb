```python
import requests
from bs4 import BeautifulSoup
import json
import codecs

url = 'https://www.ctbcbank.com/CTCBPortalWeb/toPage?id=TW_RB_CM_service_000007'   #中國信託商銀
res = requests.get(url)
soup = BeautifulSoup(res.text)
#print soup
target = soup.select('table .style01-1')
target2 = soup.select('.eventgraybox02 p')

Q = []
A = []
for question in target:
    Question = question.text
    if(len(Question)==0):
        pass
    else:
        QUESTION = Question.replace('Q：'.decode('utf8'),'')
        Q.append(QUESTION)

for answer in target2:
    Answer = answer.text
#    print len(Answer)
#    print Answer
    if(len(Answer)==0):
        pass
    else:
        ANSWER = ''.join(Answer.split()).replace('A：'.decode('utf8'),'')\
        .replace('可可以的'.decode('utf8'),'可以的'.decode('utf8'))\
        .replace('依險種整理如下(實際仍請依保險公司公告為準)：'.decode('utf8'),'請參考本行客服中心-常見問題-保險(實際請依保險公司公告為準)'.decode('utf8'))\
        .replace('綜定存質借'.decode('utf8'),'。綜定存質借'.decode('utf8')).replace('帳號：12碼'.decode('utf8'),'，帳號：12碼，'.decode('utf8'))\
        .replace('→立即設定'.decode('utf8'),'').replace('相關費用如下：'.decode('utf8'),'請洽各分行洽詢辦理'.decode('utf8'))\
        .replace('項目持續增加中。'.decode('utf8'),'項目持續增加中。2.手續費依不同繳費項目而異，執行交易前請確認手續費欄位。'.decode('utf8'))\
        .replace('','').replace('◆'.decode('utf8'),'').replace('●'.decode('utf8'),'')
        A.append(ANSWER)
        
j=1
del A[-54]
del A[-26]
del A[-21]
del A[-20]
del A[-14]
for all_q, all_a in zip(Q,A):
    
    y = {'bank':'ctbcbank','class':'complex','Q':all_q,'A':all_a} #json格式
        #key統一為bank:value自取  key統一為class:value統一請參考EXCEL  
        #銀行名稱統一小寫英文，取完之後打在EXCEL上
    
    json_qa = json.dumps(y,ensure_ascii=False,indent=2,) #轉檔轉碼  
    print json_qa   #print檢查一下
                                                         
    with codecs.open('C:/Users/vitacake/Desktop/ctbcbank/ctbcbank_complex_{}.json'.format(j),'wb') as fout: 
        json.dump(y , fout)  #寫入檔案，並關檔
    j += 1
```
