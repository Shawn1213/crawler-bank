# 華泰銀行----網路銀行(共31題)(完成)(Q24為表格)

```python
import requests
from bs4 import BeautifulSoup
import re
import json
import codecs

url = 'http://www.hwataibank.com.tw/hwatai/service_center/qa03.jsp' 
res = requests.get(url)
soup = BeautifulSoup(res.text)
#print soup

target = soup.select('table')[12:]
Q = []
A = []
del target[-18]     ##刪除掉重複抓取tr的答案()
for Target in target:
    for question in Target.select('td strong'):
        if ('Q' not in question.text):     
            pass
        else:
            QUESTION = re.sub('Q\d+\. ','',question.text)
            Q.append(QUESTION)
    for answer in Target.select('tr'):
        if('Q' in answer.text):      #最後記得加.text
            pass
        elif(len(answer)==3):
            pass
        elif('. ' in answer.text):
            pass
        else:
            Answer = (answer.text).replace('\r\n                            \t\t','').replace('\r\n                        ','')
            ANSWER = ''.join(Answer.split())
            A.append(ANSWER)
#            print ANSWER
#            print repr(answer)
#            print len(answer)


j = 1
del A[-13:-7]     #刪除掉重複抓取tr的答案()
for all_q,all_a in zip(Q,A):
    #print QS,AS
    y = {'bank':'hwatai','class':'netbank','Q':all_q,'A':all_a} #json格式
        #key統一為bank:value自取  key統一為class:value統一請參考EXCEL  
        #銀行名稱統一小寫英文，取完之後打在EXCEL上
    
    json_qa = json.dumps(y,ensure_ascii=False,indent=2) #轉檔轉碼  
    print json_qa   #print檢查一下
                                                         
#    with codecs.open('C:/Users/vitacake/Desktop/hwatai/netbank/hwatai_netbank_{}.json'.format(j),'wb') as fout: 
#        json.dump(y, fout)  #寫入檔案，並關檔
#    j += 1
```

# 華泰銀行-----放款業務Q&A(共8題)

```python
import requests
from bs4 import BeautifulSoup
import json
import codecs
url = 'http://www.hwataibank.com.tw/hwatai/personal/loan05.jsp'
res = requests.get(url)
soup = BeautifulSoup(res.text)
target = soup.select('table')[9:]
Q = []
A = []    
for Target in target:
    for question in Target.select('strong'):
        if(len(question.text)<5):
            pass
        else:
            ques = (question.text)       #問題
            Ques = re.findall('Q\d\：(.*)'.decode('utf8'),ques)    #去掉Q
            #print Ques[0]
            Q.append(Ques[0])
    content = Target.text
    Content = ''.join(content.split())    #去掉空白
#    print Content
#    print repr(Content)
    ans = re.findall('A\d\：(.*)'.decode('utf8'),Content)    #因為'：'為全形符號，所以要做解碼
    for Ans in ans:
        A.append(Ans)
j = 1
for all_q, all_a in zip(Q,A):
#    print all_q, all_a

    y = {'bank':'hwatai','class':'loan','Q':all_q,'A':all_a} #json格式
        #key統一為bank:value自取  key統一為class:value統一請參考EXCEL  
        #銀行名稱統一小寫英文，取完之後打在EXCEL上
    
    json_qa = json.dumps(y,ensure_ascii=False,indent=2) #轉檔轉碼  
    print json_qa   #print檢查一下
                                                         
#    with codecs.open('C:/Users/vitacake/Desktop/hwatai/loan/hwatai_loan_{}.json'.format(j),'wb') as fout: 
#        json.dump(y, fout)  #寫入檔案，並關檔
#    j += 1
```
