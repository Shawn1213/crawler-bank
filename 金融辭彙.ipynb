# 銀行金融辭彙表by金管會銀行局

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'http://www.banking.gov.tw/ch/home.jsp?id=203&parentpath=0,6&mcustomize='
res = requests.get(url)
soup = BeautifulSoup(res.text)
print soup
```

```python
target = soup.select('.page_content .whitebackground7')
form = []
for Target in target:
    form.append({
            'number':Target.select('.bicode_con')[0].text,   #編號
            'types':Target.select('.bitype_con')[0].text,   #分類
            'chname':Target.select('.bich_name_con')[0].text,   #中文詞彙
            'enname':Target.select('.bien_name_con')[0].text,   #英文詞彙
            'source':Target.select('.bisource_con')[0].text   #資料來源
        })
df = pd.DataFrame(form)
df = df[['number','types','chname','enname','source']]   #預設會按英文字母排列，這樣做可以照自己的去排序。
df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'http://www.banking.gov.tw/ch/home.jsp?id=203&parentpath=0,6&mcustomize='
form = []
for i in xrange(1,57):
    playload = {
        'id':'203',
        'contentid':'203',
        'parentpath':'0,6',
        'mcustomize':'bilingual_list.jsp',
        'qclass':'',
        'firstword':'',
        'ckeyword':'請輸入中文關鍵字',
        'ekeyword':'請輸入英文關鍵字',
        'page':'{}'.format(i),
        'pagesize':''
    }
    res = requests.post(url, data=playload)   #post
    soup = BeautifulSoup(res.text)
    #print soup
    target = soup.select('.page_content .whitebackground7')
    #form = []     #因為從外層迴圈進來後，又宣告空陣列，執行到最後會只剩最後的15筆。
    for Target in target:
        form.append({
            'number':Target.select('.bicode_con')[0].text,   #編號
            'types':Target.select('.bitype_con')[0].text,   #分類
            'chname':Target.select('.bich_name_con')[0].text,   #中文詞彙
            'enname':Target.select('.bien_name_con')[0].text,   #英文詞彙
            'source':Target.select('.bisource_con')[0].text   #資料來源
        })
df = pd.DataFrame(form)
df = df[['number','types','chname','enname','source']]
df
```
