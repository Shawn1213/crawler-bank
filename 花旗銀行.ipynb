```python
import requests
from bs4 import BeautifulSoup
import re
import json
import codecs

url = 'http://www.citibank.com.tw/global_docs/chi/cb/jfp_qa/index.htm'   #花旗銀行----網路銀行服務介紹(Q&A)
res = requests.get(url)
res.encoding = 'utf8'
soup = BeautifulSoup(res.text)
print soup
```

```python
texts=[]
content = soup.select('table')[6:13:3]   #把table15拿掉(第4題)
for Content in content:
    texts.append(Content)
content2 = soup.select('table')[19:-1:3]
for Content2 in content2:
    texts.append(Content2)

Q=[]
A=[]
for all_texts in texts:
    all_Texts = ''.join((all_texts.text).split()).replace('&nbsp','').replace('goto:立即註冊'.decode('utf8'),'')\
    .replace('goto:忘記網路使用者代號'.decode('utf8'),'').replace('goto:忘記網路登入密碼'.decode('utf8'),'')\
    .replace('goto:重設網路登入密碼'.decode('utf8'),'').replace('SMSOTP','SMS OTP').replace('iPadorAndroidPad','iPad or Android Pad')
    question = re.findall('Q:(.*)A:',all_Texts)
    for Question in question:
        #print Question
        Q.append(Question)
    answer = re.findall('A:(.*)',all_Texts)
    for Answer in answer:
        #print Answer
        A.append(Answer)

j=1
for all_q,all_a in zip(Q,A):
    
    y = {'bank':'citibank','class':'netbank','Q':all_q,'A':all_a} #json格式
        #key統一為bank:value自取  key統一為class:value統一請參考EXCEL  
        #銀行名稱統一小寫英文，取完之後打在EXCEL上
    
    json_qa = json.dumps(y,ensure_ascii=False,indent=2) #轉檔轉碼  
    print json_qa   #print檢查一下
                                                         
    with codecs.open('C:/Users/vitacake/Desktop/citibank/netbank/citibank_netbank_{}.json'.format(j),'wb') as fout: 
        json.dump(y , fout)  #寫入檔案，並關檔
    j += 1
```

```python
import requests
from bs4 import BeautifulSoup
import re
import json
import codecs

url = 'https://webatm.citibank.com.tw/faq.jsp'     #花旗銀行---網路ATM(共16題刪第6題)剩15題(json檔
res = requests.get(url)
soup = BeautifulSoup(res.text)
content = soup.select('table .q')
content2 = soup.select('table .a')
Q = []
A = []
for question in content:
    Ques = ''.join((question.text).split())
    Question = re.sub('\d+\.','',Ques)
    Q.append(Question)
for answer in content2:
    Answer = ''.join((answer.text).split())
    A.append(Answer)

j=1
del Q[5]
del A[5]
for all_q, all_a in zip(Q, A):
#    print all_q,all_a
    y = {'bank':'citibank','class':'atm','Q':all_q,'A':all_a} #json格式
        #key統一為bank:value自取  key統一為class:value統一請參考EXCEL  
        #銀行名稱統一小寫英文，取完之後打在EXCEL上
    
    json_qa = json.dumps(y,ensure_ascii=False,indent=2) #轉檔轉碼  
    print json_qa   #print檢查一下
                                                        
    with codecs.open('C:/Users/vitacake/Desktop/citibank/atm/citibank_atm_{}.json'.format(j),'wb') as fout: 
        json.dump(y , fout)  #寫入檔案，並關檔
    j += 1
```
